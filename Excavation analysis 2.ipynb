```python
import pandas as pd

df = pd.read_excel('Sample_data.xlsx')

df
```

# Calculating the average Weight of Artifact by Material

```python
Average_material_by_weight = df.groupby("Material")["Weight_Grams"].mean()

print('The average Weight of Artifact by Material:')
print(Average_material_by_weight)
```

# Identifying top researchers by number atifacts found

```python
Researcher_artifacts_count = df['Researcher'].value_counts()

print("Top Researchers by number of Artifacts found:")

print(Researcher_artifacts_count)
```

# Artifacts found over time

```python
import matplotlib.pyplot as plt

df['Date_Found'] = pd.to_datetime(df['Date_Found'])

df.set_index('Date_Found', inplace=True)

Artifacts_per_month = df.resample('M').size()

plt.figure(figsize = (10,6))

Artifacts_per_month.plot(kind = 'line', marker='o')

plt.title('Artifacts Discovery Over Time')

plt.xlabel('Date')

plt.ylabel('Number of Artifacts')

plt.grid(True)

plt.show()
```

# Average depth by location

```python
Average_depth_by_location = df.groupby('Location')['Excavation_Depth'].mean()

print('Average Excavation depth by Location')

print(Average_depth_by_location )
```

# Cleaning the data set

```python
df['Date_Found'] = pd.to_datetime(df['Date_Found'], errors = 'coerce')

df
```

```python
df['Location'] = df['Location'].str.strip()
df
```

```python
df['Excavation_Depth'] = pd.to_numeric(df['Excavation_Depth'], errors = 'coerce')

df
```

```python
df.dropna(subset = ['Date_Found', 'Location', 'Excavation_Depth'], inplace=True)

df
```

# Material Distribution by Artefact Type

```python
Material_distribution = df.pivot_table(index='Artifact_Type', columns= 'Material', values= 'Artifact_ID', aggfunc= 'count', fill_value = 0)

print("Material Distribution by Artefact Type:")

print(Material_distribution)
```

# Weight  Distribution by Mterial Type

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a violin plot using seaborn
plt.figure(figsize=(12, 8))

sns.violinplot(x='Material', y='Weight_Grams', data=df, inner='quart', palette='Set2')

plt.title('Weight Distribution by Material Type')

plt.xlabel('Material')

plt.ylabel('Weight (Grams)')

plt.grid(True)

# Show the plot
plt.show()
```
