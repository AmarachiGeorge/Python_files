# Temporal Analysis
This studeies the distribution of Artifacts over time by analyzing the dates they were found

```python
import pandas as pd

df = pd.read_excel('Archaeological_data.xlsx')

import matplotlib.pyplot as plt

df['DATE_FOUND'] = pd.to_datetime(df['DATE_FOUND'])

df.set_index ('DATE_FOUND', inplace = True)

Artifacts_per_month = df.resample('M').size()

plt.figure(figsize = (10, 6))

Artifacts_per_month.plot(kind = 'line', marker = 'o')

plt.title ('Artifacts Discovery over time')

plt.xlabel ('Date')

plt.ylabel ('Number of Artifacts')

plt.grid(True)

plt.show
```

# Spatial Analysis
Explores pattern of Artifacts distribution across different areas of the site

```python
import pandas as pd

df = pd.read_excel('Archaeological_data.xlsx')

import matplotlib.pyplot as plt

Area_counts = df['LOCATION'].value_counts()

plt.figure(figsize = (9, 5))

Area_counts.plot(kind ='bar', color ='blue')

plt.title ('Artifact Discovery Over Time')

plt.xlabel ('Excavation Area')

plt.ylabel ('Number of Artifacts ')

plt.xticks(rotation = 10)

plt.tight_layout()

plt.show
```

# Material Composition Analysis
Investigate the prevalence of different materials 

```python
import pandas as pd

import matplotlib.pyplot as plt

df = pd.read_excel('Archaeological_data.xlsx')

Material_count = df['MATERIAL'].value_counts()

plt.figure(figsize =(10, 6))

Material_count.plot (kind='bar', color='blue')

plt.title('Artifacts Material Composition')

plt.xlabel('Materail')

plt.ylabel('Number of Artifacts')

plt.xticks(rotation = 45)

plt.tight_layout()

plt.show()



```
